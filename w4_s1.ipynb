```python
# Store the K-means inertia in an empty list
```

```python
#@title Skeleton code
# Store the K-means inertia in an empty list
inertia = []
K = range(1,11)
for k in K:
 kmeans = KMeans(n_clusters=k, random_state=42)
 kmeans.fit(X)
inertia.append(kmeans.inertia_)
```

```python
# To finish the elbow method analysis,
# plot the change in inertia when you change the number of clusters
# you used to train the k-means clustering algorithm.

```

```python
#@title Skeleton code
# To finish the elbow method analysis, plot the change in inertia when you change the number of clusters you used to train the k-means clustering algorithm.
plt.plot(K,inertia,marker='s',c='k',lw=2)
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances / Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Store the K-means inertia in an empty list
inertia = []
K = range(1,11)

# make sure to append values to inertia for all values of k in the range
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)  # Assuming 'X' is your data
    inertia.append(kmeans.inertia_)

# To finish the elbow method analysis, plot the change in inertia when you change the number of clusters you used to train the k-means clustering algorithm.
plt.plot(K,inertia,marker='s',c='k',lw=2)
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances / Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Perform silhouette analysis following instructions in the tutorial notebook
```

```python
#@title Skeleton code
!pip install scikit-learn
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
silhouette_scores = []
for num_clusters in range(2,10):
  # initialise kmeans
  kmeans = KMeans(n_clusters=num_clusters, random_state=42)
  kmeans.fit(X)
  cluster_labels = kmeans.labels_

  # silhouette score
  silhouette_scores.append(silhouette_score(X, cluster_labels))
```

```python
# Plot your silhouette analysis result here.
plt.plot(range(2, 10), silhouette_scores, marker='o', color='b', lw=2)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal Number of Clusters')
plt.xticks(range(2, 10))  # Set the x-axis ticks to integers from 2 to 9
plt.grid()
plt.show()
```

### Q4: Perform KMeans clustering with 3 clusters
